### 3.1 ç®€å•ç¤ºä¾‹

æˆ‘ä»¬å…ˆçœ‹ä¸€ä¸ªç®€å•ä¾‹å­ï¼šå¯¹å‡½æ•° $y = 2\boldsymbol{x}^{\top}\boldsymbol{x}$ æ±‚å…³äºåˆ—å‘é‡ $\boldsymbol{x}$ çš„æ¢¯åº¦ã€‚æˆ‘ä»¬å…ˆåˆ›å»ºå˜é‡`x`ï¼Œå¹¶èµ‹åˆå€¼ã€‚

````python
x = tf.reshape(tf.Variable(range(4), dtype=tf.float32), (4,1))
x
````

å‡½æ•° $y = 2\boldsymbol{x}^{\top}\boldsymbol{x}$ å…³äº$\boldsymbol{x}$ çš„æ¢¯åº¦åº”ä¸º$4\boldsymbol{x}$ã€‚ç°åœ¨æˆ‘ä»¬æ¥éªŒè¯ä¸€ä¸‹æ±‚å‡ºæ¥çš„æ¢¯åº¦æ˜¯æ­£ç¡®çš„ã€‚

````python
with tf.GradientTape() as t:
    t.watch(x)
    y = 2 * tf.matmul(tf.transpose(x), x)

dy_dx = t.gradient(y, x)
dy_dx
````

````python
with tf.GradientTape(persistent=True) as g:
    g.watch(x)
    y = x * x
    z = y * y  # x^ 4 
    dz_dx = g.gradient(z, x)  
    dy_dx = g.gradient(y, x)  
dz_dx,dy_dx
````

## 3.2 å¯¹Pythonæ§åˆ¶æµæ±‚æ¢¯åº¦

å³ä½¿å‡½æ•°çš„è®¡ç®—å›¾åŒ…å«äº†Pythonçš„æ§åˆ¶æµï¼ˆå¦‚æ¡ä»¶å’Œå¾ªç¯æ§åˆ¶ï¼‰ï¼Œæˆ‘ä»¬ä¹Ÿæœ‰å¯èƒ½å¯¹å˜é‡æ±‚æ¢¯åº¦ã€‚

````python
def f(a):
    b = a * 2
    # ä¸æ–­å¯¹ b * 2
    while tf.norm(b) < 1000: 
        b = b * 2
    if tf.reduce_sum(b) > 0:
        c = b
    else:
        c = 100 * b
    return c
````
